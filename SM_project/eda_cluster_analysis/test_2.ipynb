## Specifiche del dataset Health Care Australia (1977-1978)

Informazioni generali:

- Il dataset si basa sull'Australian Health Survey condotto tra il 1977 e il 1978.
- Contiene dati su 5190 individui.
- Lo scopo del sondaggio era quello di raccogliere informazioni sulla salute e l'utilizzo dei servizi sanitari da parte della popolazione australiana.

Variabili:

- Sesso: 1 se femmina, 0 se maschio
- Età: Età in anni divisa per 100 (misurata come punto medio di 10 fasce d'età da 15-19 anni a 65-69 anni, con 70 o più codificati come 72)
- Età quadrata: Età al quadrato
- Reddito: Reddito annuale in dollari australiani diviso per 1000 (misurato come punto medio di intervalli codificati Nil, 200, 200-1000, 1001-, 2001, 3001-, 4001-, 5001-, 6001-, 7001-, 8001-10000, 10001-12000 12001-14000, con 14001- trattato come 15000)
- Levyplus: 1 se coperto da un'assicurazione sanitaria privata per paziente privato in ospedale pubblico (con medico di scelta), 0 in caso contrario
- Freepor: 1 se coperto dal governo perché basso reddito, immigrato recente, disoccupato, 0 in caso contrario
- Freerepa: 1 se coperto gratuitamente dal governo a causa di pensione di vecchiaia o invalidità, o perché invalido veterano o famiglia di veterano deceduto, 0 in caso contrario
- Malattia: Numero di malattie nelle ultime 2 settimane con 5 o più codificate come 5
- Giorni di attività ridotta: Numero di giorni di attività ridotta nelle ultime due settimane a causa di malattia o infortunio
- Punteggio hscore: Punteggio del questionario di salute generale usando il metodo di Goldberg. Un punteggio alto indica una cattiva salute.
- Chcond1: 1 se condizione cronica(e) ma non limitata nell'attività, 0 in caso contrario
- Chcond2: 1 se condizione cronica(e) e limitata nell'attività, 0 in caso contrario
- Doctorco: Numero di consulti con un medico o uno specialista nelle ultime 2 settimane
- Nondocco: Numero di consulti con professionisti sanitari non medici (farmacista, ottico, fisioterapista, assistente sociale, infermiere distrettuale di comunità, chiropodista o chiropratico) nelle ultime 2 settimane
- Hospadmi: Numero di ricoveri in ospedale, ospedale psichiatrico, casa di cura o convalescenza negli ultimi 12 mesi (fino a 5 o più ricoveri che vengono codificati come 5)
- Giorni di ricovero: Numero di notti in un ospedale, ecc. durante il ricovero più recente: preso, ove opportuno, come punto medio degli intervalli 1, 2, 3, 4, 5, 6, 7, 8-14, 15-30, 31-60, 61-79 con 80 o più ricoveri codificati come 80. Se nessun ricovero negli ultimi 12 mesi allora è uguale a zero.
- Medicina: Numero totale di farmaci prescritti e non prescritti utilizzati negli ultimi 2 giorni
- Prescrib: Numero totale di farmaci prescritti utilizzati negli ultimi 2 giorni
- Nonprescr: Numero totale di farmaci non prescritti utilizzati negli ultimi 2 giorni

## Levy in Australia Health Care in 70s

Healthcare in Australia during 1977-1978 was in a transitional phase, with a major focus on a universal health insurance system called Medibank. Here's a breakdown of the key points:

**Medibank**

- Introduction: Medibank was a government-funded universal health insurance scheme introduced in 1975. It aimed to provide all Australians with access to essential healthcare services regardless of income.
- How it Worked: Medibank covered basic hospital and medical costs. Patients could choose to be treated in a public hospital as a 'public patient' for free, or they could utilize private doctors/hospitals and have 85% of standard fees covered.
- Changes in 1976-78: There were several modifications to Medibank during this period. One significant change was the introduction of a levy on income to help fund the program.

**Structure of the Healthcare System**

- Federal and State Responsibilities: The Australian healthcare system has a federal structure, with responsibilities split between federal and state/territory governments. The federal government was primarily responsible for funding (like Medibank), while states and territories managed hospitals and other healthcare services.
- Public Hospitals: Public hospitals provided free or low-cost care as part of Medibank, offering essential medical treatment.
- Private Hospitals and Doctors: Australians could also choose private doctors and hospitals but had to pay part of the costs themselves, with Medibank covering a portion.

**Challenges and Issues**

- Costs: The cost of running Medibank was a significant concern for the government during this period.
- Equity: There were debates about whether Medibank provided truly equitable access for all Australians and concerns about longer waiting times in the public system.

**Important Note**: The healthcare landscape in Australia has continued to evolve since the 1970s. The current system, known as Medicare, still has a strong foundation in universal healthcare principles.

so, what if people were without private insurance and without coverage by government?

In Australia during 1977-1978, the situation for people without private insurance and government coverage was complex:

- Limited Options: Essentially, those without any form of health insurance had very limited options for accessing affordable healthcare.

- Public Hospitals: They could still rely on the public hospital system. However, as a non-Medibank patient (without any coverage), they were likely to face:

 - Out-of-pocket costs: They would be charged for any treatment or services received in the public hospital. These costs could be significant.
 - Lower Priority: Public hospitals prioritized Medibank patients and those who could contribute to their costs. People without coverage often faced longer wait times for non-emergency procedures.

- Charity: Some people might have relied on charitable organizations or the goodwill of doctors for basic medical care. However, this was not a reliable or sustainable solution.

Why Would Someone Be Without Coverage?

- Choice: A small number of people might have consciously decided against both government and private health insurance, perhaps due to ideological reasons or a strong belief in their ability to manage their own healthcare costs.
- Financial Hardship: For some, it would have been an issue of affordability. They might not have been able to pay for private insurance, and if their income was above a certain threshold, they wouldn't have qualified for free Medibank coverage.
- Ineligibility: Certain groups, such as some temporary residents or visitors, may not have been eligible for Medibank.

Consequences

The lack of health insurance coverage left these individuals vulnerable.  They bore the full financial risk of any healthcare they required, which could lead to:

- Medical Debt: Unpaid medical bills could lead to significant debt.
- Delaying Treatment: People might delay or avoid seeking necessary care due to the fear of costs, potentially worsening existing health conditions.

In Australia, a levy on income to help fund the Medibank healthcare system was introduced in October 1976. This levy was set at 2.5% of a taxpayer's taxable income.

 The levy on income to fund Medibank was in place during the 1977-1978 period. It was introduced in October of 1976.

# What is the Kruskal-Wallis Test? 

The Kruskal-Wallis test is a non-parametric method used to determine whether there are statistically significant differences between the medians of three or more independent groups. It is an extension of the Mann-Whitney U test, which is used to compare the medians of two independent groups.

The Kruskal-Wallis test is used when the assumptions of parametric tests, such as the one-way ANOVA, are not met. These assumptions include normality of the data and homogeneity of variances between groups. The Kruskal-Wallis test does not require these assumptions to be met, making it a more robust method for comparing groups with non-normally distributed data or unequal variances.

- Type of test: Non-parametric statistical test. This means it doesn't rely on assumptions about the underlying distribution of your data (e.g., normal distribution).
- Purpose: To determine if there are statistically significant differences in the medians between three or more independent groups.
- Parametric Counterpart: Similar to a one-way Analysis of Variance (ANOVA), but used when the assumptions of ANOVA are not met.

**When to Use the Kruskal-Wallis Test**:

- Your data is not normally distributed. ANOVA requires data to be normally distributed, while the Kruskal-Wallis doesn't have this restriction.
- Your data is ordinal. This means your data can be ranked (e.g., ratings like "poor," "fair," "good").
- You have three or more independent groups. The test is designed to compare multiple groups.

**How the Kruskal-Wallis Test Works**:

1. Ranking:
    - All the data from the different groups are pooled together and ranked from smallest to largest.
    - In cases of ties, the average rank is assigned.

2. Calculating the Test Statistic (H):
    - The sum of ranks for each group is calculated.
    - The H statistic is computed based on the sum of ranks, the number of observations in each group, and the total number of observations.

3. Determining Significance:
    - The H statistic is compared to a chi-square distribution table with degrees of freedom equal to the number of groups minus one.
    - If the H statistic is larger than the critical value from the table, the null hypothesis is rejected.

**Interpreting the Results**

- Null Hypothesis: The medians of all the groups are equal.
- Rejecting the Null Hypothesis: There's at least one group whose median is significantly different from the others. However, it doesn't tell you which specific groups are different. To find those, you'll need to use post-hoc tests (e.g., Dunn's test).

**Assumptions and Limitations**:
- The Kruskal-Wallis test assumes that the samples are randomly selected and independent.
- The test is sensitive to unequal sample sizes, so it's best when group sizes are similar.

    - Poisson Regression or Negative Binomial Regression: These are specifically designed for count data like 'doctorco.' They are more appropriate than linear regression if you're interested in modeling the rate of doctor consultations and avoid transforming your target variable.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import MiniBatchKMeans
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
# data = pd.read_csv('path_to_your_dataset.csv')

# Exclude the 'doctorco', 'agesq', and 'constant' column from the features used for clustering
features = data.columns.difference(['doctorco', 'agesq', 'constant'])

# Standardize the features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[features])

# Determine the range of k values for the elbow method
k_values = range(1, 40)

# Initialize the list to store the inertia values
inertia_mini = []

for k in k_values:
    # Using MiniBatchKMeans for more efficiency
    minibatch_kmeans = MiniBatchKMeans(n_clusters=k, random_state=0, batch_size=100)
    minibatch_kmeans.fit(data_scaled)
    inertia_mini.append(minibatch_kmeans.inertia_)

# Plotting the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia_mini, marker='o')
plt.title('The Elbow Method with MiniBatchKMeans')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid(True)
plt.show()

# Finding the optimal number of clusters
# Calculate the second derivative (rate of change of the inertia's rate of change)
second_derivative = np.diff(inertia_mini, 2)
optimal_clusters = np.argmin(second_derivative) + 2  # Adding 2 because np.diff reduces the array size by 1 twice

print(f"Optimal number of clusters: {optimal_clusters}")

# Run the MiniBatchKMeans with the optimal number of clusters
minibatch_kmeans_optimal = MiniBatchKMeans(n_clusters=optimal_clusters, random_state=0, batch_size=100)
cluster_labels = minibatch_kmeans_optimal.fit_predict(data_scaled)

# Add the cluster labels to your original dataframe
data['cluster'] = cluster_labels

# Analyze the clusters
print(data['cluster'].value_counts())

# Save the clustered data to a new CSV file for further analysis
data.to_csv('clustered_data.csv', index=False)

# Inspecting the centroids of the clusters
centroids = minibatch_kmeans_optimal.cluster_centers_
# If you scaled your data, reverse the scaling to interpret the centroids
centroids_original_scale = scaler.inverse_transform(centroids)
centroid_df = pd.DataFrame(centroids_original_scale, columns=features)
print(centroid_df)
```

## Cross-Analysis with 'doctorco' 

For the cross-analysis with doctorco, we'll want to examine the relationship between the cluster assignments and the number of doctor consultations. This will help to identify if there are any patterns or trends in healthcare utilization among different clusters.

Steps:
1. Aggregate doctorco by Cluster: Compute the average number of doctor consultations for each cluster. This will give you an idea of which clusters have higher or lower healthcare utilization.

2. Statistical Analysis: Perform statistical tests, if appropriate, to determine if the differences in doctorco across clusters are statistically significant.

3. Visual Comparison: Create visualizations such as bar charts or box plots to compare doctorco across clusters.

```python
#calculate the average 'doctorco' for each cluster
average_doctorco_per_cluster = data.groupby('cluster')['doctorco'].mean()

# Print the average 'doctorco' for each cluster
print(average_doctorco_per_cluster)

# Perform an ANOVA test if you want to test for statistical significance
from scipy import stats

# Perform the ANOVA test
f_val, p_val = stats.f_oneway(data['doctorco'][data['cluster'] == 0],
                              data['doctorco'][data['cluster'] == 1],
                              data['doctorco'][data['cluster'] == 2],
                              data['doctorco'][data['cluster'] == 3],
                              data['doctorco'][data['cluster'] == 4],
                              data['doctorco'][data['cluster'] == 5])

# Output the results of the ANOVA test
print("ANOVA test results: F =", f_val, ", P =", p_val)

# Visualize the data
import seaborn as sns

# Create a boxplot to see the distribution of 'doctorco' across clusters
sns.boxplot(x='cluster', y='doctorco', data=data)
plt.show()
```
